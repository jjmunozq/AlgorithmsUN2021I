##Money Change Again


```python
%%writefile mca.py

moneditas = (1, 3, 4)
dinero = int(input())


def cambio(dinero, moneditas):
	minimo,minimo[0] = [float("inf")]*(dinero + 1),0
	valorj=1
	while valorj<=dinero:
		for moneda in moneditas:
			if valorj >= moneda:
				cantidadMonedas = minimo[valorj - moneda] + 1
				if cantidadMonedas < minimo[valorj]:
					minimo[valorj] = cantidadMonedas
		valorj+=1
	return minimo[dinero]

print(cambio(dinero, moneditas))
```

##Primitive Calculator

```python
%%writefile pc.py

def minOperations(n):
    mns = [float("inf")]*(n + 1)
    mns[0: 2] = 0, 0
    atmse = [0] * (n+1)
    for i in range(2, n + 1):
        if i % 3 != 0:
            rxide_3 = float("inf")
        else:
            rxide_3 = mns[int(i / 3)]
        if i % 2 != 0:
            rxide_2 = float("inf")
        else:
            rxide_2 = mns[int(i / 2)]
        mns[i] = min(mns[i - 1], rxide_2, rxide_3) + 1
        if mns[i] == rxide_3 + 1:
            atmse [i] = int(i / 3)
            continue
        if mns[i] == rxide_2 + 1:
            atmse[i] = int(i / 2)
            continue
        if mns[i] == mns[i - 1] + 1:
            atmse[i] = i - 1
    return mns, atmse


def LCS(a,b):
    m = len(a)
    n = len(b)
    T = [[float('-inf') ] * (n + 1) for _ in range(m + 1)]
    for i in range(m + 1):
        T[i][0] = 0
    for j in range(n + 1):
        T[0][j] = 0
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i - 1] == b[j - 1]:
                T[i][j] = T[i - 1][j - 1] + 1
            else:
                T[i][j] = max(T[i - 1][j], T[i][j - 1])
    ##print(T)
    return T[m][n]

n = int(input())
mns, atmse = minOperations(n)
result = mns[n]
print(result)

camino = [n]
i = n
while i > 1:
    anterior = atmse[i]
    camino.append(anterior)
    i = anterior
camino.sort()
for number in camino:
    print(number, end = ' ')
```

##Edit Distance

```python
%%writefile ed.cpp

#include <iostream>
#include <algorithm>
#include <vector>
#include <string>

using std::string;
using std::vector;

int ed(const string &str1, const string &str2) {
  
  	int a = str1.size();
	int b = str2.size();
	vector<vector<int> > disp(a + 1, vector<int>(b + 1, 0));
	for(int y = 0; y <= a; ++y) {
        disp[y][0] = y;
    }
    for(int z = 0; z <= b; ++z) {
        disp[0][z] = z;
    }
	int cal_cost = 0;
	for(int i = 1; i <= a; ++i) {
		for(int j = 1; j <= b; ++j) {
			if(str1[i-1] == str2[j-1])
				cal_cost = 0;
			else
				cal_cost = 1;
			disp[i][j] = std::min(std::min(disp[i-1][j],disp[i][j-1]) + 1, disp[i-1][j-1] + cal_cost);
		}
	}
	return disp[a][b];
}

int main() {
  string str1;
  string str2;
  std::cin >> str1 >> str2;
  std::cout << ed(str1, str2) << std::endl;
  return 0;
}
```
