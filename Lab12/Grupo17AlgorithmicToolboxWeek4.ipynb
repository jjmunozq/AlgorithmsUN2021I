##Binary Search


```python
%%writefile bs.py

import sys

def binary_search(a, x):
	left, right = 0, len(a)-1
	s=left
	if a[0]==x:
			return(0)
	if a[right]==x:
			return(right)
	while x!=a[s]:

		if (a[left]<x and a[right]<x) or (a[left]>x and a[right]>x):
			return(-1)
		if x>a[s]:
			left=s
			s=int((right+left)/2)
			if s==left:
				return(-1)
		else:
			if x<a[s]:
				right=s
				s=int((right+left)/2)
				if s==right:
					return(-1)
	return(s)
    

def linear_search(a, x):
    for i in range(len(a)):
        if a[i] == x:
            return i
    return -1

if __name__ == '__main__':
    input = input()
    data = list(map(int, input.split()))
    n = data[0]
    m = data[n + 1]
    a = data[1 : n + 1]
    for x in data[n + 2:]:
    	print(binary_search(a,x),end = ' ')
```

##Majority Element

```python
%%writefile me.py

def get_majority_element(left, right, a):
    
    if left+1==right:
        return a[left]
    elif left+2==right:
        return a[left]

    m = (left+right)//2

    c1 = 0
    c2 = 0

    le = get_majority_element(left, m, a)
    ri = get_majority_element(m, right, a)

    for i in a[left:right]:
        if i == le:
            c1 = c1 + 1
        elif i == ri:
            c2 = c2 + 1

    if c1>(right-left)//2 and le != -1:
        return le
    elif c2>(right-left)//2 and ri != -1:
        return ri
    else: 
        return -1

n = int(input())
a = list(map(int, input().split()))

if get_majority_element(0, n, a) != -1:
    print(1)
else:
    print(0)
```

##Improving Quick Sort

```python
%%writefile iqs.py

def QuickSort3(A, l, r):
    if r <= l:
        return
    j, i = Partition3(A, l, r)
    QuickSort3(A, l, j)
    QuickSort3(A, i, r)


def Partition3(A, l, r):
    x = A[r]
    i = l - 1
    p = l - 1
    j = r
    q = r

    while True:
        while A[i + 1] < x:
            i = i + 1
        i = i + 1
        while A[j - 1] > x:
            j = j - 1
        j = j - 1
        if i >= j:
            break
        A[i], A[j] = A[j], A[i]
        if A[i] == x:
            p = p + 1
            A[i], A[p] = A[p] , A[i]
        if A[j] == x:
            q = q - 1
            A[j], A[q] = A[q], A[j]

    A[i], A[r] = A[r], A[i]

    j = i - 1
    k = l
    while k <= p:
        A[k], A[j] = A[j], A[k]
        j = j - 1
        k = k + 1

    i = i + 1
    k = r - 1
    while k >= q:
        A[k], A[i] = A[i], A[k]
        i = i + 1
        k = k - 1

    return j, i


n = int(input())
ipt = input()
a = list(map(int, ipt.split()))
QuickSort3(a, 0, n - 1)
for e in a:
    print(e, end = ' ')
```
