##6-1: Maximum Amount of Gold

```python
%%writefile mag.py

def pesomaximo(capacity, items):
    n = len(items)
    T = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            T[i][w] = T[i - 1][w]
            if w >= items[i - 1]:
                weight = T[i - 1][w - items[i - 1]] + items[i - 1]
                if weight > T[i][w]:
                    T[i][w] = weight
    #print(T)
    return T[n][capacity]


def laoptima(T, capacity, items):
    n = len(items)
    i = n
    w = capacity
    used_items = []
    while i > 0 and w > 0:
        if T[i][w] == T[i - 1][w]:
            i = i - 1
        else:
            used_items.append(i - 1)
            w = w - items[i - 1]
            i = i - 1
    return used_items


kimberly = input()
W, n = map(int, kimberly.split())
kimberly = input()
weights = list(map(int, kimberly.split()))
print(pesomaximo(W, weights))
```

##Partitioning Souvenirs

```python
%%writefile ps.py
import numpy

def partition3(u,n):
    u_ = sum(u)
    r = 0
    if n<3: 
        r = 0
    elif u_ % 3 != 0:
        r = 0
    else:
        W = u_//3

        count = 0 
        value = numpy.zeros((W+1, n+1))
        for i in range(1, W+1):
            for j in range(1, n+1):
                value[i][j] = value[i][j-1]
                if u[j-1]<=i:
                    temp = value[i-u[j-1]][j-1] + u[j-1]
                    if temp > value[i][j]:
                        value[i][j] = temp
                if value[i][j] == W:
                    count += 1
        if count < 3:
            r = 0
        else:
            r = 1

    return r

n = int(input())
u = list(map(int, input().split()))
print(partition3(u,n))
```
