##Money Change

```python
%%writefile mc.py

def get_change(m):
    i=0
    while(m>=0):
        m=m-10
        i=i+1
    m=m+10
    i=i-1
    if m==0:
        return(i)
    while(m>=0):
        m=m-5
        i=i+1
    m=m+5
    i=i-1
    if m==0:
        return(i)
    while(m>=0):
        m=m-1
        i=i+1
    i=i-1
    return(i)

if __name__ == '__main__':
    m = int(input())
    print(get_change(m))
```

##Car Fueling

```python
%%writefile cf.py
import sys

def compute_min_refills(distance, tank, stops):
    contador=0
    i=0
    anterior=0
    while i<len(stops):
        if stops[i]-anterior>tank:
            contador=contador+1
            i=i-1
            anterior=stops[i]
            if stops[i+1]-anterior>tank:
                return -1
            
        i=i+1
    return contador

if __name__ == '__main__':
    d = int(input())
    m = int(input())
    n = int(input())
    stops = list(map(int, input().split()))
    if d!=stops[len(stops)-1]:
        stops.append(d)
    print(compute_min_refills(d, m, stops))
```

##Maximum Number of Prizes

```python
%%writefile mnp.py

import math
import sys
def suma(n):
    return((n*(n+1))/2)
    

def optimal_summands(n):
    n1=math.ceil(((8*n+1)**0.5-1)/2)
    summands=[]
    for x in range(1,n1+1):
        summands.append(x)
    try:
        summands.remove(suma(n1)-n)
    except:
        pass
    return summands

if __name__ == '__main__':
    #input = sys.stdin.read()
    n = int(input())
    summands = optimal_summands(n)
    print(len(summands))
    for x in summands:
        print(x, end=' ')
```
